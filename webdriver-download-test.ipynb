```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

DRIVER_PATH = r"C:\Users\alber\miniconda3\envs\metaglass-env\lib\site-packages\chromedriver_autoinstaller\142\chromedriver.exe"

opts = Options()
opts.add_argument("--headless=new")
service = Service(executable_path=DRIVER_PATH)

with webdriver.Chrome(service=service, options=opts) as d:
    d.get("https://example.com")
    print("Título:", d.title)

print("ChromeDriver OK!")
```
